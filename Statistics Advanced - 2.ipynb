Question 1: What is hypothesis testing in statistics?

Answer:

Hypothesis testing is a statistical method used to make decisions or draw conclusions about a population using sample data.

It helps determine whether there is enough evidence to support a claim about a dataset.
It starts with a statement (hypothesis) and uses sample data to test it.

It is based on probability and helps in scientific decision-making.
The main goal is to check whether observed results are due to:

Random chance, or
A real effect in the population.
It is widely used in:
Medical research (drug effectiveness testing)
Business analysis (customer behavior)
Machine learning model evaluation


Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:

Null Hypothesis (H₀):
Assumes there is no effect, no difference, or no relationship.
It is the default assumption in hypothesis testing.

Example: “Average marks = 50”
It is the hypothesis we try to disprove.
Alternative Hypothesis (H₁ or Ha):
Assumes there is an effect, difference, or relationship.
It represents the research claim.
Example: “Average marks ≠ 50”

Difference:
H₀ → assumes nothing is happening
H₁ → assumes something is happening
Testing always focuses on whether there is enough evidence to reject H₀.


Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:

The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.
It defines the threshold for making a decision in hypothesis testing.

Common values:
0.05 (5%) → standard level
0.01 (1%) → strict level

Role in decision making:
If p-value ≤ α → Reject H₀
If p-value > α → Fail to reject H₀
It controls the risk of Type I error (false positive).

Lower α means:
More strict test
Fewer false positives
Higher α means:
More flexible test
Higher risk of error


Question 4: What are Type I and Type II errors? Give examples of each.

Answer:

Type I Error (False Positive)
Occurs when we reject H₀ even though it is true.
It means detecting something that does not exist.

Example:
A medical test shows a healthy person has a disease.

Probability = α (significance level)
Type II Error (False Negative)
Occurs when we fail to reject H₀ even though it is false.

It means missing a real effect.
Example:
A test says a sick person is healthy.

Summary:
Type I → false alarm
Type II → missed detection


Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

Answer:

Z-test:
Used when:
Sample size is large (n ≥ 30)
Population standard deviation is known
Based on normal distribution
More accurate for large datasets
Example: factory quality testing

T-test:
Used when:
Sample size is small (n < 30)
Population standard deviation is unknown
Based on t-distribution
More variability in results
Example: small clinical experiments

Key differences:
Z-test → large sample + known variance
T-test → small sample + unknown variance

```python
#7
import numpy as np
from scipy import stats

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis:
# H0: population mean = 50
# H1: population mean ≠ 50

mu = 50
sigma = 1

mean = np.mean(sample_data)
n = len(sample_data)

z = (mean - mu) / (sigma / np.sqrt(n))
p = 2 * (1 - stats.norm.cdf(abs(z)))

print("Sample Mean:", mean)
print("Z-score:", z)
print("P-value:", p)

if p < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```

```python
#8
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate normal distribution data
data = np.random.normal(50, 5, 100)

# Mean and confidence interval
mean = np.mean(data)
sem = stats.sem(data)

ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot histogram
plt.hist(data, bins=20, edgecolor='black')
plt.title("Normal Distribution")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
#9
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

# Generate sample data
data = np.random.normal(50, 10, 100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

print("Sample Z-scores:", z_scores[:10])

# Plot histogram
plt.hist(z_scores, bins=20, edgecolor='black')
plt.title("Z-score Standardized Data")
plt.xlabel("Z-score values")
plt.ylabel("Frequency")
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.
Answer:

Theory:
Z-score measures how far a value is from the mean in terms of standard deviations.

Formula:
Z = (X − μ) / σ

Interpretation:
Z = 0 → exactly equal to mean
Z > 0 → above mean
Z < 0 → below mean
Z = ±3 → extreme values (outliers)

Importance:

Helps in detecting outliers
Used in machine learning preprocessing
Standardizes different datasets
Useful in probability calculations
