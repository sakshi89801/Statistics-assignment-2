Q1. What is a Random Variable in Probability Theory?

A random variable is a function that assigns a numerical value to each possible outcome of a random experiment.

It converts outcomes into numerical form, enabling mathematical analysis.

Random variables are denoted by capital letters like
X,Y,Z.

Example: Tossing a coin — let X=1 if heads, X=0 if tails

Key points:

It can take different values based on chance.

It is measurable and associated with probabilities.

It is classified into two main types — discrete and continuous.



Q2: What are the types of random variables?

Random variables are broadly classified into:

Discrete Random Variables Can take countable or finite number of values.

Examples: Number of heads in 3 coin tosses, number of students present in a class.

They are associated with Probability Mass Functions  which give probabilities of specific outcomes.

Continuous Random Variables  Can take uncountably infinite values within a given range.

Examples: Height of students, time taken to complete a task, temperature in a city.

They are associated with Probability Density Functions  and probabilities are calculated over intervals.

Mixed Random Variables (less common) have both discrete and continuous components.

Example: Insurance claims where some claims are zero (discrete) but others vary continuously.


Q3: Explain the difference between discrete and continuous distributions.

A probability distribution describes how probabilities are assigned to values of a random variable. The main difference lies in the type of random variable involved:

Discrete Distributions

Use Probability Mass Function (PMF) to assign probabilities to each possible outcome.

The sum of all probabilities = 1.

Example: Binomial, Poisson distributions.


Continuous Distributions

Defined for continuous random variables.

Use Probability Density Function (PDF) where probabilities are given over intervals.

The probability of the variable taking any exact value is zero  only ranges have probabilities.

Example: Normal, Exponential distributions.


Q4: What is a binomial distribution, and how is it used in probability?

The Binomial distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials (each with the same probability of success).

Formula:
P(X = r) = nCr pr (1-p) n-r, r = 0, 1, 2, 3..

Where,

n = Total number of trials
r = Number of successes
p = Probability of success

Conditions for Binomial distribution:

Fixed number of trials
n

Each trial has only two possible outcomes — success or failure

Trials are independent

Probability of success
p remains constant

Example: If a coin is tossed 5 times, the probability of getting exactly 3 heads can be calculated using binomial distribution.

Uses:

Quality control defective items in a batch

Analysis yes/no responses

Testing pass/fail outcomes

Q7: What is the significance of confidence intervals in statistical analysis?
Definition:

A confidence interval (CI) is a range of values, derived from sample statistics, that is likely to contain the true population parameter (mean, proportion, etc.) with a specified level of confidence.

Key Points:

Confidence Level:

The probability that the interval contains the true parameter. Common levels: 90%, 95%, 99%.

Example: A 95% CI means that if we take 100 samples, about 95 of them will contain the true mean.

where:

X = sample mean

Z α/2 = Z-value from standard normal distribution

σ = population standard deviation

n = sample size

Importance in Statistics:

Quantifies uncertainty of estimates.

Used in decision-making for research, policy, and business.

Avoids the mistake of relying solely on point estimates.


Q8: What is the concept of expected value in a probability distribution?
Definition:

The expected value (EV) of a random variable is the long-run average value it would take over an infinite number of trials. It represents the “center” or “mean” of a probability distribution.

Properties:

Linearity:
E[aX+b]=aE[X]+b

Additivity:
E[X+Y]=E[X]+E[Y] if X and Y are independent.

Interpretation:
EV is not always a possible value of X (e.g., expected dice roll = 3.5).

Importance:

Decision Making: Used in economics, finance, and insurance to choose strategies with the best long-term payoff.

Risk Assessment: Helps quantify potential outcomes.




```python
# 9
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)
print("Mean:", np.mean(data))
print("Std Dev:", np.std(data, ddof=1))

plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution")
plt.show()
```

```python
#10
import numpy as np
from scipy import stats

sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
         235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean = np.mean(sales)
se = np.std(sales, ddof=1) / np.sqrt(len(sales))
ci = stats.t.interval(0.95, len(sales)-1, loc=mean, scale=se)

print("Mean Sales:", mean)
print("95% CI:", ci)
```
