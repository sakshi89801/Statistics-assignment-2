Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.
Answer:

Statistics is broadly divided into descriptive statistics and inferential statistics, both serving different purposes in data analysis.

1. Descriptive Statistics

Descriptive statistics refers to the methods used to organize, summarize, and present data in a meaningful way. It simply describes the data that is already available without making any predictions or generalizations.

It uses tools like:

Mean, median, mode
Tables and graphs (bar charts, histograms, pie charts)
Standard deviation and variance

Example:
If a teacher calculates the average marks of 50 students in a class, that is descriptive statistics. It only summarizes the class performance.

2. Inferential Statistics

Inferential statistics refers to the techniques used to analyze a sample of data and make conclusions or predictions about a larger population. It uses probability to estimate unknown values.

It includes:

Hypothesis testing
Confidence intervals
Regression and correlation analysis

Example:
If a company surveys 1,000 customers and predicts the satisfaction level of all customers in India, it is inferential statistics.

Key Difference:
Descriptive = describes existing data
Inferential = makes predictions about population


Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.
Answer:
1. Sampling

Sampling is the process of selecting a smaller group (sample) from a large population to study and draw conclusions. It is used because studying the entire population is often difficult, costly, or time-consuming.

2. Random Sampling

Random sampling is a method where each member of the population has an equal chance of being selected.

Features:
No bias in selection
Simple and fair method
Every individual has equal probability

Example:
Selecting 10 students randomly from a class using a lottery system.

3. Stratified Sampling

Stratified sampling is a method where the population is first divided into different groups (strata) based on a characteristic, and then samples are taken from each group.

Features:
Ensures representation of all groups
More accurate than random sampling
Reduces sampling error

Example:
In a school, students are divided into boys and girls, and samples are taken from both groups equally.

Difference:
Random sampling = completely random selection
Stratified sampling = grouped selection for better representation


Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.
Answer:

Measures of central tendency are statistical tools used to represent a dataset with a single central value.

1. Mean

Mean is the average of all values in a dataset. It is calculated by dividing the sum of all values by the number of values.

2. Median

Median is the middle value when data is arranged in ascending or descending order. If there are even numbers, the average of the two middle values is taken.

3. Mode

Mode is the value that occurs most frequently in a dataset.

Importance:
Helps in summarizing large datasets
Useful for comparing data easily
Widely used in business, economics, and research
Helps in decision-making and forecasting


Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?
Answer:
1. Skewness

Skewness measures the asymmetry of a data distribution.

If data is symmetric → no skew
If not symmetric → skew exists
Types:
Positive skew (right skew)
Negative skew (left skew)
Positive Skew:

A positive skew means:

Most values are concentrated on the left (lower values)
Tail is stretched toward the right (higher values)
Few extremely high values exist

Example:
Income distribution where most people earn less, but a few earn very high salaries.

2. Kurtosis

Kurtosis measures the shape of the distribution curve, especially the peak and tails.

High kurtosis → sharp peak, heavy tails
Low kurtosis → flat distribution
Importance:
Helps understand data distribution shape
Useful in risk analysis and financial modeling

```python
#6
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation:", correlation)
```
